# Imports

```julia
using EpistemicNetworkAnalysis
```

# Load Data

```julia
data = ena_dataset("shakespeare.data")
display(data)
```

# ENA Model

```julia
conversations = [:Play, :Act]
units = [:Play, :Speaker]
codes = [
    :Love,
    :Death,
    :Honor,
    :Men,
    :Women
]

rotation = MeansRotation(:Play, "Romeo and Juliet", "Hamlet")

myENA = ENAModel(data, codes, conversations, units, rotateBy=rotation)
display(myENA)
```

# Plotting

```julia
p = plot(myENA)
display(p)
```
