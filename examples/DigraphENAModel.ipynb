# Digraph ENA Model

## Parameters

The `DigraphENAModel` constructor takes the same parameters as the `ENAModel` constructor. The only difference is that, because this model is directed, one can include both `X_Y` and `Y_X` orderings of a pair of codes `X` and `Y` with the `relationshipFilter` optional parameter.

```julia
conversations = [:Play, :Act]
units = [:Play, :Speaker]
codes = [
    :Love,
    :Death,
    :Honor,
    :Men,
    :Women
]

rotation = MeansRotation(:Play, "Romeo and Juliet", "Hamlet")

myENA = DigraphENAModel(data, codes, conversations, units, rotateBy=rotation)
display(myENA)
```

# Plotting

```julia
p = plot(myENA)
display(p)
```
